Develop a program to Load a dataset with at least two numerical columns (e.g., Iris, Titanic). Plot a scatter 
plot of two variables and calculate their Pearson correlation coefficient. Write a program to compute the 
covariance and correlation matrix for a dataset. Visualize the correlation matrix using a heatmap to know 
which variables have strong positive/negative correlations

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('iris_dataset (1).csv')
df.info()
```

```python
df.head(5)
```

```python
df1='sepal_length'
df2='sepal_width'
```

```python
plt.figure(figsize=(5,5))
sns.scatterplot(x=df[df1],y=df[df2],hue=df['species'],palette='inferno')
plt.title(f"Scatter plot of {df1} v/s {df2}")
plt.xlabel(df1)
plt.ylabel(df2)
plt.show()
```

```python
pear=df[df1].corr(df[df2])
print(f"Pearson corelation coefficient between {df1} and {df2}:",pear)
```

```python
cov_matrix=df[[df1,df2]].cov()
print("\nCovariance Matrix:")
print(cov_matrix)
```

```python
corr_matrix=df.drop(columns=['species']).corr()
print("\nCorrelation Matrix:")
print(corr_matrix)
```

```python
plt.figure(figsize=(5,3))
sns.heatmap(corr_matrix,annot=True,cmap='plasma',fmt='.1f')
plt.title('Corelation Matrix of Heatmap')
plt.show()
```
