3.DataFrame objects

A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.

```python
p_dic={
  'Weight':pd.Series([2,4,5,6,8],index=['Rock','Thor','Thanos','Dhanoosh','Raja']),
  'DOY':pd.Series([1980,1999,1979,1992,2000],index=['Rock','Thor','Thanos','Dhanoosh','Raja'],name='year'),
  'Job':pd.Series(['Wrestler','Hero','Rescurer','Youtuber','Legend'],index=['Rock','Thor','Thanos','Dhanoosh','Raja'])
}
```

```python
p_dic
```

```python
bio=pd.DataFrame(p_dic)
bio
```

```python
f=pd.DataFrame(
    bio,
    columns=['Weight','DOY',],
    index=['Rock','Raja']
)
```

```python
f
```

Passing all the values constructor

```python
values=[
    [1985,'Racing',68],
    [1945,'Fighter',89],
    [1992,0,np.nan,112]
]
d3=pd.DataFrame(
    values,
    columns=['DOY','Children','Hobby','Weight'],
    index=['S','B','C']
)
d3
```

To Specify missing values,you can use np.nan

Multi-index

```python
data=pd.DataFrame(
    {
        ('public','Doy'):
        {('Paris','Alice'):1985,('Paris','bob'):1984,('London','Charles'):1982},
        ('public','hobby'):
        {('Paris','Alice'):'Biking',('Paris','bob'):'Dancing'},
        ('private','Weight'):
        {('Paris','Alice'):85,('Paris','bob'):np.nan,('London','Charles'):0},
        ('private','Height'):
        {('Paris','Angel'):85,('Paris','bob'):100,('London','Charles'):0},
    }
)
data
```
