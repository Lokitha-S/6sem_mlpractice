3.DataFrame objects

A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.

```python
p_dic={
  'Weight':pd.Series([2,4,5,6,8],index=['Rock','Thor','Thanos','Dhanoosh','Raja']),
  'DOY':pd.Series([1980,1999,1979,1992,2000],index=['Rock','Thor','Thanos','Dhanoosh','Raja'],name='year'),
  'Job':pd.Series(['Wrestler','Hero','Rescurer','Youtuber','Legend'],index=['Rock','Thor','Thanos','Dhanoosh','Raja'])
}
```
